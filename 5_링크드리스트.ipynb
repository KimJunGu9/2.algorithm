# **1. 링크드리스트(Linked List)**

* 떨어진 곳에 존재하는 데이터를 화살표로 연결해서 관리하는 구조
* C언어에서는 중요한 자료구조지만, 파이썬에서는 리스트 타입이 링크드 리스트 역할도 모두 지원
* 데이터의 삽입과 삭제가 매우 빠름

### **1-1. 링크드리스트 용어**
* 노드(node) : 데이터 저장 단위(데이터, 포인터)로 구성
* 포인터(pointer) : 각 노드 안에서 다음이나 이전의 노드와의 연결정보를 가지고 있는 공간

1.png

```python
# 링크드리스트 예

# 파이썬에서는 링크드 리스트를 구현할 때 클래스를 주로 활용

class Node:
  def __init__(self, data):
    self.data = data   # 데이터
    self.next = None   # 포인터를 가리키는 변수. 일단 비워줌
```

```python
# Node 객체와 Node 객체를 연결하기 (포인터 활용)

node1 = Node(1)  # 1을 data에 넣음
node2 = Node(2)  # 2를 data에 넣음
print(node1)
print(node2)
node1.next = node2
head = node1
```

### **1-2. 링크드 리스트로 데이터 추가하기**

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

  def add(self, data):
    node = head # 시작점
    while node.next:  # while문 옆에는 True & False만 들어갈 수 있는데, 데이터가 있으면 True이다. 0이거나 None이면 멈춘다.(node.next = Node(data)줄로 내려간다)
      node = node.next

    node.next = Node(data)
```

```python
node1 = Node(1)
head = node1
for index in range(2, 11):
  node1.add(index)
```

# **2. 객체지향 프로그래밍으로 링크드 리스트 구현**

```python
class Node:
  def __init__(self, data, next=None):
    self.data = data
    self.next = next
```

```python
class NodeMgmt:
  def __init__(self, data): # data에 숫자를 넣어주면
    self.head = Node(data) # 그 숫자에 의해서 Node객체를 class Node:에 의해서 만들어낸다.
                          # 그리고 변수를 self.head에 저장하면, 일단 가장 처음에  저장되는 node객체가 head가 가리킴


  def add(self, data):
    if self.head == None: # self의 head가 존재하지않는다면(None이면)
      self.head = Node(data) # self.head에다가 node객체를 처음부터 만들어서 넣어줌
                            # 방어코드(여기로 들어올 가능성 거의 없다)
    else:
      node = self.head # head(제일 처음 데이터)를 node에 넣어서
      while node.next:  # 다음 자료가 있는지 확인한다.
        node = node.next
      node.next = Node(data)



  def node_print(self): # 잘 저장됐는지 print해보기 # self만 있으니 매개변수는 없다.
    node = self.head # 처음데이터를 node에 넣고
    while node: # node객체가 있는지 판단한 후에, 있다면 내려가서 node.data를 찍어주고
      print(node.data)
      node = node.next # 그리고 다음 자료로 넘어가기



  def delete(self, data):
    # head에 노드가 없을 경우
    if self.head == None:
      print('head에 노드가 없습니다')
      return


    # 지우려는 값이 head에 있는 경우
    if self.head.data == data:
      print('지우려는 값이 head에 있습니다')
      temp = self.head
      self.head = self.head.next # head값이 지워지고 다음데이터가 head로 들어감.
      del temp

    # 지우려는 값이 head에 없는 경우
    else:
      print('지우려는 값이 head에 없습니다')
      node = self.head
      while node.next:  # 다음데이터가 있니? 있으면 내려가
        if node.next.data == data:
          temp = node.next
          node.next = node.next.next
          del temp
          return
        else:
          node = node.next
```
